```python
from langchain.prompts import PromptTemplate
```

## 3. First Prompt Template

```python
# Example 1: Summary Generator

from langchain.prompts import PromptTemplate

# Define template
template = "Write a short summary about {topic}"

# Create PromptTemplate object
prompt = PromptTemplate(
    input_variables=["topic"],
    template=template
)

# Format the prompt
output = prompt.format(topic="Artificial Intelligence")

print(output)
```

## 4.Second Prompt Template

```python
# Example 2: Question Answering

template = "Answer the following question: {question}"

prompt = PromptTemplate(
    input_variables=["question"],
    template=template
)

output = prompt.format(question="What is Machine Learning?")

print(output)
```

## 4.Third Prompt Template

```python
# Example 3: Email Generator

template = "Write a professional email for {purpose}"

prompt = PromptTemplate(
    input_variables=["purpose"],
    template=template
)

output = prompt.format(purpose="requesting internship extension")

print(output)
```

## 5.Fourth Prompt Template

```python
# Example 4: Multiple Inputs

template = "Explain {topic} in {level} level"

prompt = PromptTemplate(
    input_variables=["topic", "level"],
    template=template
)

output = prompt.format(topic="AI", level="beginner")

print(output)
```
